
Dane z Apple Watch mogą być bogate w informacje, szczególnie jeśli obejmują statystyki dotyczące zdrowia, aktywności fizycznej i snu. Z poniższymi danymi można stworzyć ciekawe i intuicyjne wykresy oraz statystyki. Oto kilka przykładów:

1. Kroki i Aktywność Fizyczna
Dane:
Liczba kroków
Pokonany dystans
Spalone kalorie
Czas aktywności
Minuty intensywnej aktywności
Wykresy i statystyki:
Dzienna liczba kroków: Wykres słupkowy pokazujący kroki w poszczególnych dniach.
Trend tygodniowy/miesięczny: Linia trendu przedstawiająca zmiany w liczbie kroków, dystansie lub spalonych kaloriach.
Porównanie aktywności w różnych porach dnia: Wykres kołowy lub histogram.
Cel aktywności vs. rzeczywista aktywność: Wskaźnik postępu w procentach.
2. Tętno
Dane:
Średnie tętno spoczynkowe
Średnie tętno podczas aktywności
Maksymalne i minimalne tętno
Wartości tętna w czasie rzeczywistym
Wykresy i statystyki:
Średnie dzienne tętno spoczynkowe: Wykres liniowy pokazujący trendy w czasie.
Tętno podczas ćwiczeń: Wykres obszarowy przedstawiający zmienność w trakcie aktywności.
Porównanie tętna z intensywnością aktywności: Wykres rozrzutu.
Histogram tętna: Podział tętna na przedziały (np. spoczynkowe, umiarkowane, intensywne).
3. Sen
Dane:
Czas snu
Fazy snu (lekki, głęboki, REM)
Godzina zaśnięcia i obudzenia
Wykresy i statystyki:
Podsumowanie czasu snu: Wykres słupkowy pokazujący, ile godzin użytkownik spał każdego dnia.
Proporcje faz snu: Wykres kołowy przedstawiający podział na fazy.
Jakość snu w czasie: Linia trendu pokazująca zmiany w długości i jakości snu w ciągu tygodni/miesięcy.
Godziny zasypiania i budzenia: Wykres punktowy lub histogram, aby zidentyfikować regularność.
4. Ćwiczenia i Trening
Dane:
Typ aktywności (np. bieganie, rower, joga)
Czas trwania treningu
Spalone kalorie
Średnie tempo i dystans
Wykresy i statystyki:
Porównanie typów aktywności: Wykres słupkowy dla czasu trwania lub spalonych kalorii w zależności od aktywności.
Mapa aktywności GPS: Wizualizacja trasy na mapie (dla aktywności takich jak bieganie czy rower).
Trend czasu trwania treningu: Wykres liniowy pokazujący postępy w tygodniach/miesiącach.
Porównanie spalonych kalorii na różne aktywności: Wykres kołowy.
5. Strefy Czasu Siedzenia
Dane:
Liczba godzin siedzenia
Czas, w którym użytkownik wstaje i rusza się
Wykresy i statystyki:
Czas siedzenia vs. aktywność: Wykres obszarowy pokazujący proporcje.
Przerwy w siedzeniu: Histogram liczby wstań w ciągu dnia.
Porównanie dni roboczych i weekendów: Wykres słupkowy.
6. Zdrowie Ogólne
Dane:
Poziom natlenienia krwi (SpO2)
VO2 Max
Zmienność rytmu serca (HRV)
Temperaturę ciała (jeśli dostępna)
Wykresy i statystyki:
Trendy poziomu SpO2: Wykres liniowy pokazujący zmiany w czasie.
Zmienność rytmu serca (HRV): Wykres słupkowy z podziałem na dni.
Porównanie VO2 Max: Trendy w kondycji fizycznej.
Związek temperatury ciała i aktywności: Wykres rozrzutu.
7. Personalizacja i Statystyki Długoterminowe
Rekordy osobiste: Największa liczba kroków, najdłuższy trening itp.
Cele i osiągnięcia: Wskaźniki postępu w wyznaczonych celach.
Analiza cykliczności: Dni najbardziej i najmniej aktywne.
Narzędzia do wizualizacji w Streamlit
Aby stworzyć takie wykresy w Streamlit, możesz użyć bibliotek takich jak:

Matplotlib lub Seaborn: Tworzenie klasycznych wykresów.
Plotly: Interaktywne wykresy.
Altair: Intuicyjne i szybkie wizualizacje.
Mapy: Do tras GPS.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import streamlit as st
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
df.head(72)
```

```python
df["Calories_per_reading"] = df.groupby("user_id")["Calories"].diff().fillna(df["Calories"])
```

```python
df["Steps_per_reading"] = df.groupby("user_id")["Steps"].diff().fillna(df["Steps"])
```

```python
df["Distance_per_reading"] = df.groupby("user_id")["Distance"].diff().fillna(df["Distance"])
```

```python
calories_by_activity = df.groupby("activity_trimmed")["Calories_per_reading"].sum().reset_index()

st.subheader("Wykres spalonych kalorii według typu aktywności")
fig, ax = plt.subplots()
ax.bar(calories_by_activity["activity_trimmed"], calories_by_activity["Calories_per_reading"], color="skyblue")
ax.set_xlabel("Typ aktywności")
ax.set_ylabel("Spalone kalorie")
ax.set_title("Kalorie według aktywności")
st.pyplot(fig)
```
