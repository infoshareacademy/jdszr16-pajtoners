```python
data = pd.read_csv('data_for_weka_aw.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data = data.drop("Unnamed: 0", axis=1)
```

```python
filtered_data = data[['age', 'height', 'weight']]

unique_users = filtered_data.drop_duplicates()

num_unique_users = unique_users.shape[0]
num_unique_users
```

```python
unique_users['user_id'] = range(1, unique_users.shape[0] + 1)

data = data.merge(unique_users, on=['age', 'height', 'weight'], how='right')

columns = ['user_id'] + [col for col in data.columns if col != 'user_id']
data = data[columns]

data.head()
```

```python
phrase_to_del = ["Applewatch", "_LE"]

for phrase in phrase_to_del:
    data.columns = data.columns.str.replace(phrase, "", regex=True)
    data.columns = data.columns.str.replace('.', '', regex=False)

data.head()
```
