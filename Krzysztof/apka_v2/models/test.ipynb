```python
import pandas as pd
import pickle
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
data_path = 'test_data.csv'
df = pd.read_csv(data_path)
```

```python
scaler_path = "scaler_new.pkl"
with open(scaler_path, 'rb') as f:
    scaler = pickle.load(f)

kmeans_path = "kmeans_new.pkl"
with open(kmeans_path, 'rb') as f:
    kmeans = pickle.load(f)
    
pca_path = "pca_new.pkl"
with open(pca_path, 'rb') as f:
    pca = pickle.load(f)
```

```python
kmeans_features = ['Heart', 'RestingHeartrate', 'NormalizedHeartrate', 'CorrelationHeartrateSteps', 'age', 'weight']
numeric_df = df[kmeans_features]

scaled_data = scaler.transform(numeric_df)
reduced_data = pca.transform(scaled_data)
```

```python
numeric_df
```

```python
y_pred = kmeans.predict(reduced_data)
```

```python
y_pred
```

```python
df["predictions"] = y_pred
```

```python
df.head()
```

```python
silhouette_avg = silhouette_score(scaled_data, y_pred)
```
